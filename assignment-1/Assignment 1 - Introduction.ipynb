```python
import numpy as np
import numpy.polynomial.polynomial as poly
import math
import matplotlib.pyplot as plt
import warnings
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import linear_model

# params for full-screen graphs
plt.rcParams['figure.figsize'] = (20.0, 10.0)

# supress warnings
warnings.filterwarnings('ignore')
```

### Part a

```python
def quadratic_data_gererator(x):
    return -0.25*np.square(x)+5

x = np.array([-0.15,-0.05, -0.047, -0.05,0.065,0.062, 0.063])
y = quadratic_data_gererator(x)
```

```python
def fit(x, y, deg):
    coef = np.polyfit(x, y, deg)
    ffit = np.poly1d(coef)
    x_new = np.linspace(x[0], x[-1])
    y_new = ffit(x_new)
    return x_new, y_new
    
```

```python
plt.figure(1)

plt.subplot(131)
x_new, y_new = fit(x, y, 1)
plt.plot(x, y, 'bo', x_new, y_new, '--k')
plt.title('Underfitting')
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='both', which='both', bottom='off', top='off', 
                labelbottom='off', left='off', labelleft='off')

plt.subplot(132)
x_new, y_new = fit(x, y, 2)
plt.plot(x, y, 'bo', x_new, y_new, '--k')
plt.title('Appropriate Capacity')
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='both', which='both', bottom='off', top='off', 
                labelbottom='off', left='off', labelleft='off')

plt.subplot(133)
x_new, y_new = fit(x, y, 9)
plt.plot(x, y, 'bo', x_new, y_new, '--k')
plt.title('Overfitting')
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='both', which='both', bottom='off', top='off', 
                labelbottom='off', left='off', labelleft='off')

plt.show()
```

### part b

```python
def fit_weight_decay(x, y, deg, alpha):
    X = x.reshape(-1,1)
    model = make_pipeline(PolynomialFeatures(deg), Ridge(alpha=alpha, normalize=True, solver='lsqr'))
    model.fit(X, y)
    x_new = np.linspace(x[0], x[-1]).reshape(-1, 1)
    y_new = model.predict(x_new)
    return x_new, y_new
```

```python
plt.figure(2)

plt.subplot(131)
x_new, y_new = fit_weight_decay(x, y, 1, 10**10)
plt.plot(x, y, 'bo', x_new, y_new, '--k')
plt.title('Underfitting\n(Excessive )')
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='both', which='both', bottom='off', top='off', 
                labelbottom='off', left='off', labelleft='off')

plt.subplot(132)
x_new, y_new = fit_weight_decay(x, y, 2, 10**-1)
plt.plot(x, y, 'bo', x_new, y_new, '--k')
plt.title('Appropriate Weight Decay\n(Medium )')
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='both', which='both', bottom='off', top='off', 
                labelbottom='off', left='off', labelleft='off')

plt.subplot(133)
x_new, y_new = fit_weight_decay(x, y, 9, 10**-10)
plt.plot(x, y, 'bo', x_new, y_new, '--k')
plt.title('Overfitting \n(  0)')
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='both', which='both', bottom='off', top='off', 
                labelbottom='off', left='off', labelleft='off')



plt.show()
```
